```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
meta_file = 'raw_data/metadata.csv'
meta_df = pd.read_csv(meta_file)
meta_df.columns.tolist()
```

```python
meta_df = meta_df.dropna(subset=['pubmed_id', 'sha']).reset_index(drop=True)
meta_pmids = np.array(meta_df['pubmed_id'].values, dtype='int').tolist()
meta_pmids = set([str(i) for i in meta_pmids])
len(meta_pmids)
```

```python
chem_gene_ixns_relation = 'raw_data/KG/chem_gene_ixns_relation.csv'
chem_df = pd.read_csv(chem_gene_ixns_relation, delimiter='\t')
chem_df
```

```python
chem_pmids = set(chem_df.pmids.tolist())
chem_pmids & meta_pmids
```

```python
actions = chem_df.InteractionActions.tolist()
```

```python
fst_act = []
snd_act = []
for act in actions:
    acts = act.split('|')
    for a in acts:
        a = a.split('^')
    fst_act.append(a[0])
    snd_act.append(a[1])
```

```python
print(Counter(fst_act))
print(Counter(snd_act))
```
